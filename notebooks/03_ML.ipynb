# Bemorga dori tavsiya qilish tizimi

Kerakli kutubxonalarni chaqirib olamiz

```python
import pandas as pd # Dataset bilan ishlash
import numpy as np # Vektor hisob kitoblar
from sklearn.model_selection import train_test_split # Datasetni train va test qisimlarga ajratish
from sklearn.tree import DecisionTreeClassifier # Decision Tree algorthmi
from sklearn.metrics import classification_report # Modelni baholash
from sklearn.metrics import confusion_matrix # Modelni baholash
from sklearn.model_selection import cross_val_predict
```

Tayyorlab olgan datasetimizni yuklab olamiz

```python
df = pd.read_csv("../data/processed/drug200_transformed.csv")
df.head()
```

`X` va `y` larni ajratib olamiz

```python
X = df[['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K']].values
```

```python
y = df['Drug'].values
```

Train va Test setlarga ajratamiz

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=40)
```

## ML

```python
tree_model = DecisionTreeClassifier()

tree_model.fit(X_train, y_train)
```

**Confusion Matrix** - Modelni baholash

```python
y_predicted = tree_model.predict(X_test)
```

```python
confusion_matrix(y_test, y_predicted)
```

**Classification Report** - modelni baholash

```python
print(classification_report(y_test, y_predicted))
```

**Cross-validation**

```python
predict = cross_val_predict(estimator = tree_model, X=X, y=y, cv=5)
print("Classification Report: \n", classification_report(y, predict))
```
